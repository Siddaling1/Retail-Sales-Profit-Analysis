# 1.What are the earliest and latest order dates in the dataset?

```python
Latest_date=df['Order_Date'].max()
Earliest_date=df['Order_Date'].min()
print("Latest_date is :" , Latest_date)
print("Earliest_date is :" , Earliest_date)
```

# 2.Which year recorded the highest total sales and profit?

```python
df['Order_Date']=pd.to_datetime(df["Order_Date"])
df["year"]=df['Order_Date'].dt.year
```

```python
df["Total_Sales"]=df['Sales']*df['Quantity']
```

```python
df.head()
```

```python
High_sales_and_Profits=df.groupby('year')[['Total_Sales' , 'Profit']].sum()
High_sales_and_Profits
```

# 3.What is the monthly trend of sales and profit over time?

```python
df['Months']=df['Order_Date'].dt.strftime("%b")
```

```python
df.head()
```

```python
months=df.groupby('Months')[['Total_Sales' , 'Profit']].sum()
months
```

```python
import matplotlib.pyplot as plt
months.plot(kind='bar')
plt.xlabel("Months")
plt.ylabel("Total Sales and Profits")
plt.title("Total Sales and Profit By Months")
plt.show()
```

# 4.Which product categories and sub-categories generate the highest sales?

```python
Category=df.groupby(['Category' , 'Sub-Category'])['Total_Sales'].sum()
Category
```

# 5.Which region contributes the most to total profit?

```python
Region=df.groupby('Region')['Profit'].sum()
Region
Region.sort_values(ascending=False)
```

# 6.What is the average discount offered across different product categories?

```python
Average_Discount =df.groupby('Category')['Discount'].mean()
Average_Discount
```

# 7.Which customers have placed the most orders?

```python
Most_orders=df.groupby('Customer_Name')['Row_ID'].sum()
Most_orders.sort_values(ascending=False).head(5)
```

# 8.What is the correlation between sales, profit, and discount?

```python
Correlation=df[['Total_Sales','Profit', 'Discount']].corr()
Correlation
```

# 9.What are the top 10 most profitable products?

```python
Profitable_Product=df.groupby('Product_Name')['Profit'].sum()
Profitable_Product.sort_values(ascending=False).head(10)
```

# 10.Identify loss-making products (where profit < 0).

```python
loss_products = df[df['Profit'] < 0]
loss_products[['Product_Name' ,'Profit']]
```
